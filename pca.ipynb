```python
#This library is used for data manipulation and analysis.
import pandas as pd
#this library is efficient operations on arrays, which are essential for numerical computations.
import numpy as np
# This library is used for creating various visualizations like plots, charts, and histograms.
import matplotlib.pyplot as plt
```

```python
#This line imports the breast_cancer dataset from the sklearn.datasets module.
from sklearn.datasets import load_breast_cancer
#This line calls the load_breast_cancer function and assigns the returned value to a variable named breast_cancer.
breast_cancer = load_breast_cancer()
```

```python
#The breast_cancer is a Bunch object containing information about the Wisconsin Breast Cancer dataset.
breast_cancer
```

```python
#the breast_cancer.keys() function returns a dictionary view of the keys in the breast_cancer dataset.
#These keys correspond to the different attributes of the dataset.
breast_cancer.keys()
```

```python
breast_cancer.data
```

```python
#features names function returns list of strings containing the names of the features (measurements).
breast_cancer.feature_names
```

```python
#The target function returns a NumPy array of shape (569- shape ) indicating whether a sample is malignant or benign
breast_cancer.target
```

```python
#pd.DataFrame: This creates a new pandas DataFrame object.
#in the second line numpy.c_ (concatenate columns) is called to combine two arrays into a single array. i.e..
#breast_cancer.data and breast_cancer.target
#The 3rd line sets the column names for the DataFrame.
#list(breast_cancer.feature_names): This converts the feature_names attribute into a list.
#target: This adds a new column named "target" at the end
df = pd.DataFrame(
np.c_[breast_cancer.data, breast_cancer.target],
columns = [list(breast_cancer.feature_names)+ ['target']]
)
df.head()
```

```python
#this function returns statistical measures of the dataframe
df.describe()
```

```python
#X is a pandas DataFrame containing the feature matrix, which has a shape of (569, 30), indicating 569 samples and 30 features.
#Y is a NumPy array containing the target labels, which has a shape of (569), indicating 569 samples.
X = df
Y = breast_cancer.target
```

```python
#this function returns the header part of the dataframe X.
X.head()
```

```python
#import the libraries from sklearn
#The 1st function performs Min-Max scaling on numerical data.
#It rescales each feature in the dataset to a range between a specified minimum and maximum value, typically between 0 and 1
from sklearn.preprocessing import MinMaxScaler
#This function performs standardization on numerical data. It centers and scales the data by subtracting the mean and dividing by the
#standard deviation for each feature. This results in a dataset with a mean of 0 and a standard deviation of 1.
from sklearn.preprocessing import StandardScaler
```

```python
#This line creates a new instance of the StandardScaler from sklearn.preprocessing.
scaler = StandardScaler()
#This line fits the scaler to the data in the DataFrame X. During the fitting process, the scaler calculates the mean and
#standard deviation for each feature in the dataset.
scaler.fit(X)
```

```python
#This command calls the transform method of the fitted scaler object. It takes the DataFrame X containing the features as input.
scaled_data = scaler.transform(X)
```

```python
scaled_data
```

```python
#This line imports the train_test_split function, which is used to split data into training and testing sets.
from sklearn.model_selection import train_test_split
#This line splits the data using train_test_split and assigns the resulting subsets to four variables:
#X_train: This variable will contain a DataFrame holding the training features (after standardization).
#X_test: This variable will contain a DataFrame holding the testing features (after standardization).
#Y_train: This variable will contain a NumPy array holding the training target labels.
#Y_test: This variable will contain a NumPy array holding the testing target labels.
#The argument test size specifies the proportion of data to be included in the testing set.
#This argument sets a random seed for splitting the data. This ensures reproducibility, meaning if you run the code again with
#the same random seed, you'll get the same split of data into training and testing sets.
X_train, X_test, Y_train, Y_test = train_test_split(scaled_data, Y,test_size = 0.4, random_state = 25)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(X)
```

```python
X_pca = pca.transform (X)
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = breast_cancer.target, cmap = "viridis")
plt.xlabel ("PC1")
plt.ylabel("PC2")
plt.title("PCA on Breast Cancer")
plt.colorbar(ticks=[0,1,2], label = 'breast cancer class')
plt.show()
```
